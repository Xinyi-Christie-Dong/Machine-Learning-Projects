```python
#导入数据

import pandas as pd

# 定义文件路径
file_path = '../Data/3.2原数据.xlsx'

# 读取Excel文件前19行
data = pd.read_excel(file_path, usecols=range(19))

# 显示数据
print(data)
```

```python
#描述性统计

import pandas as pd
from scipy.stats import skew, kurtosis

# 创建一个空的DataFrame来存储统计结果
stats_df = pd.DataFrame(columns=['Variable', 'Mean', 'Std Dev', 'Skewness', 'Kurtosis', 'Min', 'Max', 'Autocorrelation'])

# 计算并添加数据到DataFrame
for i in range(2, 18):  # 第三列到第十八列的索引
    column = data.iloc[:, i]
    
    # 计算统计指标
    mean = column.mean()
    std_dev = column.std()
    skewness = skew(column)
    kurt = kurtosis(column)
    min_val = column.min()
    max_val = column.max()
    autocorr = column.autocorr()

    # 从原始数据获取列名
    column_name = data.columns[i]
    
    # 将结果添加到DataFrame
    stats_df = stats_df.append({
        'Variable': column_name,
        'Mean': mean,
        'Std Dev': std_dev,
        'Skewness': skewness,
        'Kurtosis': kurt,
        'Min': min_val,
        'Max': max_val,
        'Autocorrelation': autocorr
    }, ignore_index=True)

# 打印表格
print(stats_df)

#输出excel表格
output_excel_path = '../Results/3.2描述性统计.xlsx'
stats_df.to_excel(output_excel_path, index=False)
```

```python
#单变量回归

import pandas as pd
import statsmodels.api as sm

# 用于存储回归结果的DataFrame
results_df = pd.DataFrame(columns=['Variable', 'Coefficient', 'Std Error', 't Value', 'P Value','R-squared'])

# 执行回归分析并添加结果
for i in range(2, 19):  # 第三列到第十九列的索引
    # 定义自变量和因变量
    X = data.iloc[:, [i]]  # 第i列是自变量
    X = sm.add_constant(X)  # 添加常数项
    y = data.iloc[:, 1]  # 第二列作为因变量

    # 构建并拟合模型
    model = sm.OLS(y, X)
    results = model.fit()
    
    # 输出结果
    print(f"自变量 {i-1}:")
    print(results.summary())
    print("\n")

    # 将结果添加到DataFrame
    for param_name, value in results.params.items():
        results_df = results_df.append({
            'Variable': param_name,
            'Coefficient': value,
            'Std Error': results.bse[param_name],
            't Value': results.tvalues[param_name],
            'P Value': results.pvalues[param_name],
            'R-squared': results.rsquared
        }, ignore_index=True)
        
# 导出为Excel
output_excel_path = '../Results/3.2单变量回归结果.xlsx'
results_df.to_excel(output_excel_path, index=False)
```
